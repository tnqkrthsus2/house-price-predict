```python
import numpy as np
import pandas as pd
```

```python
a = pd.read_excel('A(m2,소비자심리,미분양).xlsx')
```

```python
df = pd.merge(df, a, on = '계약년월', how = 'left')
df
```

```python
print(df[df['GNI'].isnull()])
```

```python
from sklearn import linear_model
lin_reg = linear_model.LinearRegression()
```

```python
df.dtypes
```

```python
df = df.astype({'평균코스피지수':'float64'})
```

GNI 결측치 제거

```python
X = df.dropna(axis=0)[['소비자물가지수','평균코스피지수']] 
y = df.dropna(axis=0)['GNI'] 
lin_reg_model = lin_reg.fit(X, y)
```

```python
df.pop('주택담보대출금리')
```

```python
df.pop('대출평균')
```

```python
y_pred = lin_reg_model.predict(df.loc[:, ['소비자물가지수','평균코스피지수']])
#x변수에 결측치있는지 확인하기
```

```python
df['GNI'] = np.where(df['GNI'].isnull(), 
                              pd.Series(y_pred.flatten()), 
                              df['GNI'])
```

GDP 결측치 제거

```python
X = df.dropna(axis=0)[['소비자물가지수','평균코스피지수']] 
y = df.dropna(axis=0)['GDP'] 
lin_reg_model = lin_reg.fit(X, y)
y_pred = lin_reg_model.predict(df.loc[:, ['소비자물가지수','평균코스피지수']])
df['GDP'] = np.where(df['GDP'].isnull(), 
                              pd.Series(y_pred.flatten()), 
                              df['GDP'])
```

GDI 결측치 제거

```python
X = df.dropna(axis=0)[['소비자물가지수','평균코스피지수']] 
y = df.dropna(axis=0)['GDI'] 
lin_reg_model = lin_reg.fit(X, y)
y_pred = lin_reg_model.predict(df.loc[:, ['소비자물가지수','평균코스피지수']])
df['GDI'] = np.where(df['GDI'].isnull(), 
                              pd.Series(y_pred.flatten()), 
                              df['GDI'])
```

결측치 행 삭제

```python
df = df[df['주차대수'].notna()]
df = df[df['분양형태'].notna()]
df = df[df['M2통화량(평잔,십억원)'].notna()]
df = df[df['주택시장소비심리지수'].notna()]
df = df[df['미분양'].notna()]
```
