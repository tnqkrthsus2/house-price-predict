```python
import seaborn as sns
import stylecloud
```

```python

for i in df['apart__name'].unique():
    try:
        okt = Okt()
        value = globals()["df_{}".format(i)]['content'].values
        value = value.tolist()
        text = str(value)[:]
        nouns = okt.nouns(text) # 명사만 추출
        dong = df['town_name'][df['apart__name']==i].iloc[0] 
        remove_nouns = ['아파트','서울','평형','건설','단지','산업','가격','지역',dong,'거래','주민','현대']
        for j in remove_nouns:
            while j in nouns:
                nouns.remove(j)
        words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

        c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
        wc = WordCloud(font_path='malgun', background_color='white',colormap='PuBu',
        width=400, height=400, scale=2.0, max_font_size=250)
        gen = wc.generate_from_frequencies(c)
        plt.figure()
        plt.imshow(gen)
        h = df['apart_code'][df['apart__name']==i].iloc[0]
        wc.to_file('wordcloud_%s.png' % h)
    except:
        print('error')
```
