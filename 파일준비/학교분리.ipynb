```python
gizun  = pd.read_csv('gizun.csv',encoding='cp949')
```

```python
gizun = gizun[['아파트코드','좌표Y','좌표X']]
```

```python
gizun.columns = ['apart_code','X','Y']
```

```python
gizun
```

```python
from haversine import haversine
```

```python
near_elements = []
near_element_distance = []
for i in gizun.index:
    minium_distance = 100
    start = (float(gizun['X'][i]),float(gizun['Y'][i]))
    for j in element.index:
        goal =  (float(element['x'][j]),float(element['y'][j]))
        distance = haversine(start, goal,'km')
        if distance < minium_distance:
            minium_distance = distance
            near_element = element['school_name'][j]
    near_elements.append(near_element)
    near_element_distance.append(minium_distance)
gizun['가까운 초등학교'] = near_elements
gizun['초등학교까지 거리'] = near_element_distance
```
