```python
import numpy as np
import pandas as pd
```

```python
apt = pd.read_csv('중구아파트.csv')
```

```python
apt
```

```python
df = apt[['아파트코드','전용면적(㎡)', '계약년월', '건축년도', '주차대수','범죄율', '건물연식',
'대규모점포_거리', '버스정류장_거리', '지하철역_거리', '학교_거리', '유치원_거리', '병원_거리',
'거래금액(만원)','주택구매력지수', 'RIR','LIR']]
```

```python
df
```

```python
df['area']=df['전용면적(㎡)'].round(-1)
```

```python
df['area']
```

```python
df.pop('전용면적(㎡)')
```

```python
df['price'] = 0
for i in df.index:
    if type(df['거래금액(만원)'][i]) == str:
        df['price'][i] = df['거래금액(만원)'][i].replace(',','')
    else: 
        df['price'][i] = df['거래금액(만원)'][i]
df        
```

```python
df = df.astype({'price':'float64'})
```

```python
df['price']=df.groupby(['계약년월','아파트코드','area'])['price'].transform('mean')
```

```python
df['아파트구분'] = df['아파트코드'].map(str) +'-' +df['area'].map(str)
```

```python
df
```

```python
df = df.drop_duplicates(subset=['아파트구분','계약년월'])
```

```python
df.pop('거래금액(만원)')
```

```python
df['date'] = df['계약년월']
df = df.set_index('계약년월')
```

```python
df
```

```python
df = df.sort_values(by=["아파트구분", "date"], ascending=[True, True]) 
```

```python
#수정해야됨
import numpy as np
def make_dataset(data, label, window_size=5):
    feature_list = []
    label_list = []
    for i in data.index:
        if data['code'][i] = data['code'][i + window_size]:
            feature_list.append(np.array(data.iloc[i:i+window_size]))
            label_list.append(np.array(label.iloc[i+window_size]))
        else 
    return np.array(feature_list), np.array(label_list)
```
